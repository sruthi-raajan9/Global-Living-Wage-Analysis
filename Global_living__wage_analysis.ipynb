```python
# clean the column names
df.columns = df.columns.str.replace('\n', ' ').str.strip()
```

```python
# create the list of columns that contain currency/money values

currency_cols = [
    'Std Family Urban (Annual)', 'Std Family Rural (Annual)',
    'Typical Fam Urban (Annual)', 'Typical Fam Rural (Annual)',
    'Single Indiv Urban (Annual)', 'Single Indiv Rural (Annual)',
    'Single Parent Urban (Annual)', 'Single Parent Rural (Annual)'
]

# Clean and convert to numeric
# Convert to string to handle 'USD' or ',' if they exist
for col in currency_cols:
    df[col] = df[col].astype(str).str.replace('USD', '', regex=False)
    df[col] = df[col].str.replace(',', '', regex=False)

# Convert to numeric
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#  Adding a representative column
df['Median(LW)'] = (df[currency_cols].median(axis=1)).round(0)
```

```python
df.head()
```

**URBAN RURAL COST GAP ANALYSIS**

```python
# Calculate the percentage difference for a standard family urban vs rural
df['U/R Gap%(std fam)'] = (((df['Std Family Urban (Annual)'] - df['Std Family Rural (Annual)']) /
                            df['Std Family Rural (Annual)']) * 100).round(2)
```

```python
# Calculate the percentage difference for a single parent urban vs rural
df['U/R Gap%(single parent)'] = (((df['Single Parent Urban (Annual)'] - df['Single Parent Rural (Annual)']) /
                            df['Single Parent Rural (Annual)']) * 100).round(2)
```

```python
# Calculate the percentage difference for a single individual urban vs rural
df['U/R Gap%(single individual)'] = (((df['Single Indiv Urban (Annual)'] - df['Single Indiv Rural (Annual)']) /
                            df['Single Indiv Rural (Annual)']) * 100).round(2)
```

```python
# Calculate the percentage difference for a Typical Fam urban vs rural
df['U/R Gap%(Typical Fam)'] = (((df['Typical Fam Urban (Annual)'] - df['Typical Fam Rural (Annual)']) /
                            df['Typical Fam Rural (Annual)']) * 100).round(2)
```

**SINGLE PARENT FINANCIAL BURDEN**

```python
# finanacial burden of a single parent in urban area by comparing standard family
df['(Sing parent burden% vs std fam(U))']=((df['Single Parent Urban (Annual)']-df['Std Family Urban (Annual)'])/df['Std Family Urban (Annual)']*100).round(2)
```

The Affordability Index

    Q. Which countries have the highest and lowest cost of survival relative to their economic status?

```python
#  Calculate the PPP Affordability Index using the Living Wage column of std family -urban
country_gdp['Aff_Index(std_fam_urban)']= (country_gdp['Std Family Urban (Annual)'] / country_gdp['GDP per capita(ppp)']).round(1)
```

```python
# check how many countries that has affordability index greater than 0.40
country_gdp.loc[country_gdp['Aff_Index(std_fam_urban)']>0.40].head()
```

```python
# sort by affordability index value
country_gdp.sort_values(by='Aff_Index(std_fam_urban)', ascending=False).head()
```

```python
country_gdp.sort_values(by='Aff_Index(std_fam_urban)', ascending=False).tail(5)
```

```python
#  Calculate the PPP Affordability(single individual urban) Index using the Living Wage column
country_gdp['Aff_Index(Sin_Indv_Urban)']= (country_gdp['Single Indiv Urban (Annual)'] / country_gdp['GDP per capita(ppp)']).round(1)
```

```python
country_gdp.loc[country_gdp['Aff_Index(Sin_Indv_Urban)']>0.40]
country_gdp.sort_values(by='Aff_Index(Sin_Indv_Urban)', ascending=False).head()
```

```python
#  Calculate the PPP Affordability(single parent urban) Index using the Living Wage column
country_gdp['Aff_Index(sin_parent_Urban)']= (country_gdp['Single Parent Urban (Annual)'] / country_gdp['GDP per capita(ppp)']).round(1)
```

```python
#  Calculate the PPP Affordability(country median) Index using  median(LW)
country_gdp['Aff_Index(country_med)']= (country_gdp['Median(LW)'] / country_gdp['GDP per capita(ppp)']).round(1)
```

```python
#  Calculate the PPP Affordability(typical family urban) Index using the living wage column
country_gdp['Aff_Index(Typ fam urban)']= (country_gdp['Typical Fam Urban (Annual)'] / country_gdp['GDP per capita(ppp)']).round(1)
```

```python
#  Filter, select columns, and sort in ascending order
affordable_countries = country_gdp.loc[
    country_gdp['Aff_Index(country_med)'] < 0.40,
    ['Country Name', 'Aff_Index(country_med)']
].sort_values(by='Aff_Index(country_med)', ascending=True)

# Display the result
print(affordable_countries)
```
